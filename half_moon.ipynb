```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.datasets import make_moons
```

```python
# plot original dataset using make_moons, here just using sample=1000 noise=0.3 which is close to the hw image
X,y=make_moons(n_samples=1000,noise=0.3,random_state=0)
plt.figure()
plt.plot(X[np.where(y==1)[0], 0], X[np.where(y==1)[0], 1], 'o')
plt.plot(X[np.where(y==0)[0], 0], X[np.where(y==0)[0], 1], 's', c = 'r')
# build a logistic classifer 
lr = LogisticRegression()
lr.fit(X, y,sample_weight=None)

#generate new dataset and learn from logistic model and predict-->classify dataset
X1,y1=make_moons(n_samples=2000,noise=0.3,random_state=0)
pre_LM = lr.predict(X1)

#plot new graph with decison boundary
plt.figure()
plt.plot(X1[np.where(pre_LM==1)[0], 0], X1[np.where(pre_LM==1)[0], 1], 'o')
plt.plot(X1[np.where(pre_LM==0)[0], 0], X1[np.where(pre_LM==0)[0], 1], 's', c = 'r')
```
